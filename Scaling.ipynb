```python
#Preamble
import sarabande
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axes_grid1 as axgrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
import astropy.io.fits as pyf
import os
import time

plt.style.use("light_mode") #custom matplotlib style sheet

sarabande.about()
```

```python
hdulist = pyf.open('data/dens_t800.fits.gz')
data = hdulist[0].data.astype(np.float64)
data -= np.mean(data)

data_32 = data[:32,:32,:32]
data_64 = data[:64,:64,:64] 
data_128 = data[:128,:128,:128]
data_256 = data

data_list = [data_32, data_64, data_128, data_256]
```

```python
#string to directory to save data into
save_dir = os.getcwd() + '/output/'

dts_3PCF, _3PCFs = [], []
for i, data_i in enumerate(data_list):
    #create measure_obj
    _3PCF_i = sarabande.measure(nPCF=3, projected=False, density_field_data = data_i, save_dir=save_dir, save_name='example', nbins=4, ell_max=1)
    
    start = time.clock()
    sarabande.calc_zeta(_3PCF_i, verbose_flag=False)
    stop = time.clock()
    
    _3PCFs.append(_3PCF_i)
    dts_3PCF.append(stop - start)
```

```python
#string to directory to save data into
save_dir = os.getcwd() + '/output/'

dts_4PCF, _4PCFs = [], []
for i, data_i in enumerate(data_list):
    #create measure_obj
    _4PCF_i = sarabande.measure(nPCF=4, projected=False, density_field_data = data_i, save_dir=save_dir, save_name='example', nbins=4, ell_max=1)
    
    start = time.clock()
    sarabande.calc_zeta(_4PCF_i, verbose_flag=False)
    stop = time.clock()
    
    _4PCFs.append(_4PCF_i)
    dts_4PCF.append(stop - start)
```

```python
#string to directory to save data into
save_dir = os.getcwd() + '/output/'

dts_3PCF_p, _3PCFs_p = [], []
for i, data_i in enumerate(data_list):
    #create measure_obj
    _3PCF_i = sarabande.measure(nPCF=3, projected=True, density_field_data = data_i[:,:,16], save_dir=save_dir, save_name='example', nbins=10, m_max=5)
    
    start = time.clock()
    sarabande.calc_zeta(_3PCF_i, verbose_flag=False)
    stop = time.clock()
    
    _3PCFs_p.append(_3PCF_i)
    dts_3PCF_p.append(stop - start)
```

```python
#string to directory to save data into
save_dir = os.getcwd() + '/output/'

dts_4PCF_p, _4PCFs_p = [], []
for i, data_i in enumerate(data_list):
    #create measure_obj
    _4PCF_i = sarabande.measure(nPCF=4, projected=True, density_field_data = data_i[:,:,16], save_dir=save_dir, save_name='example', nbins=10, m_max=5)
    
    start = time.clock()
    sarabande.calc_zeta(_4PCF_i, verbose_flag=False)
    stop = time.clock()
    
    _4PCFs_p.append(_4PCF_i)
    dts_4PCF_p.append(stop - start)
```

```python
import scipy.optimize as fitter

Ns = [32, 64, 128, 256]
N_strings = [str(N) for N in Ns]

#we have to define our model with our needed parameters
def model_NlogN(x, a, b, c):
    return a * x*np.log(c*x) + b

# def model_quad(x, a, b, c):
#     return a + b*x + c*x*x

par0    = np.array([1.0, 2.0, 1.0]) # initial guess for parameters
par, cov = fitter.curve_fit(model_NlogN, Ns, dts_4PCF_p, par0) #fitter.curve_fit takes in the model, x,y data, guess, and sigma

# par arrays contains the values of parameters. cov is the covariance matrix
# decode it now
a = par[0]
ea = np.sqrt(cov[0,0])
print('a={0:6.3f}+/-{1:5.3f}'.format(a,ea))
b = par[1]
eb = np.sqrt(cov[1,1])
print('b={0:6.3f}+/-{1:5.3f}'.format(b,eb))
c = par[2]
ec = np.sqrt(cov[2,2])
print('c={0:6.3f}+/-{1:5.3f}'.format(c,ec))

# params_3pcf_proj = [a,b,c]
params_4pcf_proj = [a,b,c]
```

## Component Wise Timing

```python
def compute_components(measure_obj):
    dt_components = []
    t0 = time.process_time()
    sarabande.create_XYZR(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.create_radial_bins(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_ft_data(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_and_save_YLMs(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.bin_spherical_harmonics(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_a_lm_coeffs(measure_obj, verbose=False, kernel_name = measure_obj.save_name)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_zeta(measure_obj, skip_prepare=True)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    percents = list(np.array(dt_components) / sum(dt_components) * 100)

    percents = [(sum(percents[:4]))] + percents[4:]
    
    return percents


def compute_components_projected(measure_obj):
    dt_components = []
    t0 = time.process_time()
    sarabande.projected_create_bins(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.projected_create_kernels(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_ft_data(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.projected_create_Cm_coeffs(measure_obj)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    t0 = time.process_time()
    sarabande.calc_zeta(measure_obj, skip_prepare=True)
    dt = time.process_time() - t0
    dt_components.append(dt)
    
    percents = list(np.array(dt_components) / sum(dt_components) * 100)

    percents = [(sum(percents[:2]))] + percents[2:]
    
    return percents

    
    
    
```
